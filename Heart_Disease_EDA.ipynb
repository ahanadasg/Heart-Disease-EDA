```python
import pandas as pd

# Load the dataset
heart_data = pd.read_csv('heart.csv')
```

```python
# View the first few rows of the dataset
print(heart_data.head())

# Check data types and non-null values
print(heart_data.info())

# Summary statistics of numerical features
print(heart_data.describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the heart disease dataset (assuming the data is in a DataFrame called 'heart_data')
# heart_data = pd.read_csv('heart_disease_dataset.csv')

# Categorical columns
categorical_columns = ['sex', 'age']

# Create subplots for each categorical column
num_cols = len(categorical_columns)
fig, axs = plt.subplots(1, num_cols, figsize=(15, 5))

# Create pie charts for each categorical column
for i, column in enumerate(categorical_columns):
    # Calculate value counts for each category
    category_counts = heart_data[column].value_counts()
    labels = category_counts.index.tolist()
    sizes = category_counts.values.tolist()

    # Plot the pie chart
    axs[i].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
    axs[i].set_title(column)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the heart disease dataset (assuming the data is in a DataFrame called 'heart_data')
# heart_data = pd.read_csv('heart_disease_dataset.csv')

# Numeric columns for the violin plot
numeric_columns = ['age', 'trestbps', 'chol', 'thalach']

# Create subplots for each numeric column
num_cols = len(numeric_columns)
fig, axs = plt.subplots(1, num_cols, figsize=(15, 5))

# Create violin plots for each numeric column
for i, column in enumerate(numeric_columns):
    sns.violinplot(x='target', y=column, data=heart_data, ax=axs[i], hue='target', split=True)
    axs[i].set_title(column)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the heart disease dataset (assuming the data is in a DataFrame called 'heart_data')
# heart_data = pd.read_csv('heart_disease_dataset.csv')

# Select numerical columns for correlation analysis
numeric_columns = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'thal']

# Compute the correlation matrix
correlation_matrix = heart_data[numeric_columns].corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the heart disease dataset (assuming the data is in a DataFrame called 'heart_data')
# heart_data = pd.read_csv('heart_disease_dataset.csv')

# Select numerical columns for the Pair Plot
numeric_columns = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'thal']

# Create the Pair Plot
sns.pairplot(heart_data[numeric_columns])
plt.suptitle("Pair Plot for Heart Disease Dataset", y=1.02)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the heart disease dataset (assuming the data is in a DataFrame called 'heart_data')
# heart_data = pd.read_csv('heart_disease_dataset.csv')

# Select two numerical columns for correlation analysis
column1 = 'age'
column2 = 'thalach'

# Create the scatter plot with a regression line
plt.figure(figsize=(8, 6))
sns.regplot(x=column1, y=column2, data=heart_data, scatter_kws={'alpha':0.5})
plt.title(f"Scatter Plot: {column1} vs. {column2}")
plt.xlabel(column1)
plt.ylabel(column2)
plt.show()
```
